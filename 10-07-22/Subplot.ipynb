```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)
x1=np.array([0,1,2,3])
y1=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x1,y1)
plt.show()  
```
