```python
import matplotlib.pyplot as plt
cartoon=["Doremon","Tom&Jerry","Ninja Hatori","Phenes&Ferb"]
Watch_Hours=[5,7,3,4]
plt.bar(cartoon,Watch_Hours)
plt.title("Cartoon watch hours per day")
plt.xlabel("Cartoon")
plt.ylabel("Watch_Hours")
plt.show()
```

```python
plt.barh(cartoon,Watch_Hours,color="hotpink")
plt.show()
```
