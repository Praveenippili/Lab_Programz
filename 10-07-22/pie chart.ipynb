```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
c=["r","b","hotpink"]
plt.pie(y,colors=c,labels=["dsp","cd","wt"])
plt.legend(title="subject")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,27,15,23])
labels=["cadboury","kitkat","5star","perk"]
plt.pie(y,startangle=90,labels=labels)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,27,15,23])
labels=["cadboury","kitkat","5star","perk"]
e=[0.1,0,0,0]
plt.pie(y,startangle=90,labels=labels,explode=e,shadow=True)
plt.show()
```
