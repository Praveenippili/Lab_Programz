```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
x=np.array([2,80])
y=np.array([3,90])
plt.plot(x,y)
plt.show()
```

```python
plt.plot([5,20],[4,90],"--")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.scatter(x,y)
plt.show()
```

```python
y=np.array([3,8,1,10,5,7])
plt.plot(y)
plt.show()
```

```python
y=np.array([3,8,1,10,5,7])
plt.plot(y,marker="o")
plt.show()
```

```python
y=np.array([3,8,1,10,5,7])
plt.plot(y,marker="o",mec="r")
plt.show()
```

```python
y=np.array([3,8,1,10,5,7])
plt.plot(y,marker="o",mec="r",color="k")
plt.show()
```

```python
y=np.array([3,8,1,10,5,7])
plt.plot(y,linewidth="10")
plt.show()
```
