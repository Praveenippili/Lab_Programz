```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x=np.array([5,15,25,35,45,55])
print(x)
x=np.array([5,15,25,35,45,55]).reshape(-1,1)
print(x)
y=np.array([5,20,14,32,22,38])
print(y)
```

```python
model=LinearRegression()
model.fit(x,y)
print(model.intercept_)
print(model.coef_)
print(model.score(x,y))
```

```python
y_pred=model.predict(x)
print(y_pred)
print(model.score(x,y_pred))
plt.scatter(x,y)
plt.plot(x,y_pred)
plt.show()
```

```python
print(model.intercept_+model.coef_*x)
```

```python
x_new=np.arange(5).reshape(-1,1)
print(x_new)
y_new=model.predict(x_new)
print(y_new)
print(model.score(x_new,y_new))
plt.scatter(x_new,y_new)
plt.plot(x_new,y_new)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
x=np.array([10,15,20,25,30])
y=np.array([5,10,20,40,80])
print(x)
X=x[:,np.newaxis]
print(X)
model=LinearRegression().fit(X,y)
yfit=model.predict(X)
plt.scatter(X,y)
plt.plot(X,yfit)
plt.show()
```

```python
data=pd.read_csv("data.csv")
plt.scatter(data["Age"],data["Salary"],c="red")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
x=data["Age"].values.reshape(-1,1)
y=data["Salary"].values
```

```python
reg=LinearRegression()
reg.fit(x,y)
print(reg.score(x,y))
print(reg.coef_)
print(reg.intercept_)
y_pred=reg.predict(x)
plt.scatter(data["Age"],data["Salary"],c="blue")
plt.plot(data["Age"],y_pred,c="red",linewidth=2)
plt.show()
print(reg.score(x,y_pred))
```
